```python
# Create a new column 'price_group' and assign the value based on car price
# Body type and Drive type analysis
import pandas_profiling
import numpy as np                                                 # Implemennts milti-dimensional array and matrices
import pandas as pd 
import matplotlib as mat               # Plotting library for Python programming language
import matplotlib.pyplot as plt
import bokeh                           # interactive visualization library for Python
import sys                             
import os
import seaborn as sns 
%matplotlib inline
cardata['price_group'] = pd.cut(cardata['price'],[0,40000,80000,600000], 
                                labels = ['<40K','40-79K','>100K'], include_lowest = True)
cardata['price_group'] = cardata['price_group'].astype(object)

(cardata['price_group'].value_counts() / len(cardata) *100).plot.bar()
plt.title("Price Range Vs Total Cars")
plt.ylabel('% of vehicles')
plt.xlabel('Price Group');
```

```python
# Create a new column 'price_group' and assign the value based on car price
# Body type and Drive type analysis
import pandas_profiling
import numpy as np                                                 
import pandas as pd 
import matplotlib as mat               
import matplotlib.pyplot as plt
import bokeh                           # interactive visualization library for Python
import sys                             
import os
import seaborn as sns 
%matplotlib inline


# Brand wise car sale in higher price segment
carprice = cardata.loc[cardata['price'] >= 200000]
(carprice['car'].value_counts() / len(carprice)*100).plot.bar()                
plt.title("Brand wise Sales Distribution in Higher Price Segment")
plt.ylabel('% of vehicles')
plt.xlabel('Brand type');
```

```python
# price v/s mileage 
cardata[cardata['price'] < 100000 ].plot.scatter(x='price', y='mileage')
plt.title("Price V/S Mileage")
plt.ylabel('Mileage')
plt.xlabel('Price');
```

```python
# production year wise count of vehicles
carYear = cardata.loc[cardata['year'] >= 2000]
carYear['year'].value_counts().sort_index().plot.line()
plt.title("Production Year VS count of vehicles")
plt.ylabel('Count of vehicles')
plt.xlabel('year');
```

```python
# Production year v/s registration
import seaborn as sns 
cc = cardata.loc[cardata['year'] >= 2000]
g = sns.catplot('year',data=cc,aspect=2.5, kind='count',hue='registration',color="g")
g.set_xticklabels(rotation=30)
```
